```python
from nupack import *
import numpy as np
import re
import os
import csv
DFT_CONCERNTRATION = 1e-1
MODEL = Model(material = 'dna', celsius = 37,sodium=1.0,magnesium=0.0,ensemble='stacking')
for j in range(50,150):
    file = './Data/rand_dna_seq_'+str(j)+'.csv'
    score_all=[]
    with open(os.path.join(file),'r') as f:
        reader = csv.reader(f)
        for i in reader:
            seq = i[0]
            seq1 = Strand(seq,name="a")
            my_complex = Complex([seq1],name="b")
            tube1 = Tube({seq1:DFT_CONCERNTRATION},complexes=SetSpec(max_size=1,include=[my_complex]),name="tube1")
            single_results = tube_analysis([tube1],model=MODEL,compute=['pfunc', 'pairs', 'mfe', 'sample', 'subopt'],options={'num_sample': 2, 'energy_gap':0.5})
            score = single_results[my_complex].mfe[0].energy
#             line = str(score) +'\n'+ str(seq)
            score_all.append(score)
    save_file='pred_score_seq'+str(j)+'.csv'
    with open(os.path.join('./pred/',save_file),'w',newline='') as f2:
        writer = csv.writer(f2)
        for z in score_all:
            writer.writerow([z])
    print(j)
```
